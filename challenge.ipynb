```python
import pandas as pd 

# Read the data
data = pd.read_csv('data.csv')
data['minute_of_hour'] = data['alarm (%H:%M:%S)'].apply(lambda x: int(x.split(':')[1]))
data.head() 
```

```python
# plot the data
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.scatter(data['day'], data['minute_of_hour'])
```

```python
# do regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(data[['day']], data['minute_of_hour'])

# plot the regression line
plt.plot(data['day'], model.predict(data[['day']]), color='red')
plt.show()
```
