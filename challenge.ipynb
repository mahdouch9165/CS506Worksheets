```python
import pandas as pd

df = pd.read_csv('data.csv')
#Rename column alarm (%H:%M:%S) to time
df.rename(columns={'alarm (%H:%M:%S)':'time'}, inplace=True)
df.head()
```

```python
df['second_of_day'] = df['time'].apply(lambda x: int(x.split(':')[0])*3600 + int(x.split(':')[1])*60 + int(x.split(':')[2]))
df.head()
```

```python
# Scatter plot
import matplotlib.pyplot as plt

plt.scatter(df['day'], df['second_of_day'])
```

```python
# log secodn of day
import numpy as np
import math

df['sqrt_second_of_day'] = df['second_of_day'].apply(lambda x: 1/(x+1))

plt.scatter(df['day'], df['sqrt_second_of_day'])
```
