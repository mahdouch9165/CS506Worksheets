## Linear Regression

a) Create and plot a dataset of 100 (Y, X) points where `Y = 1 + 0.5 * X + eps`.

```python
import numpy as np
import matplotlib.pyplot as plt

SAMPLE_SIZE = 100

true_beta = np.array([ 1 , 3 ])
x = -1.0 + 2.0 * np.random.random(SAMPLE_SIZE)
y = true_beta[0] + true_beta[1] * x + np.random.randn(SAMPLE_SIZE)
plt.plot(x,y,'ro',markersize=4)
plt.show()
```

b) Plot the least squares estimate line through the scatter plot.

```python
intercept = np.ones(np.shape(x)[0])
X = np.array([intercept, x]).T
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

x_plot = np.linspace(-1,1,50)
y_est = beta_hat[0] + beta_hat[1] * x_plot
plt.plot(x_plot, y_est,'b--', lw=2, label='Estimate')
plt.plot(x_plot, true_beta[0] + true_beta[1] * x_plot, lw=2, c='g', label='Truth')
plt.plot(x, y,'ro',markersize=4)
plt.legend()
plt.show()
```

c) How does the estimate of `beta` compare to the parameters we used to generate the data?

very close

d) Create and plot a dataset of 100 (Y, X) points where `Y = 1 + 3 * X + .5 * X^2 + eps`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of data points
n = 100

# Generating X values uniformly between -10 and 10
X = np.random.uniform(-10, 10, n)

# Generating Y values based on the given formula
# eps is a normally distributed noise
eps = np.random.normal(0, 1, n)
Y = 1 + 3 * X + 0.5 * X**2 + eps

# Plotting the dataset
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.title('Generated Dataset with Quadratic Relationship and Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

e) Plot the least squares estimate line through the scatter plot.

```python
from numpy.linalg import lstsq

A = np.vstack([X, np.ones(len(X))]).T
m, c = lstsq(A, Y, rcond=None)[0]
line_of_best_fit = m * X + c
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, line_of_best_fit, color='red', label='Least Squares Estimate Line')
plt.title('Least Squares Estimate Line through the Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# print beta
print("Beta_0: ", c)
print("Beta_1: ", m)
```

f) How does the estimate of `beta` compare to the parameters we used to generate the data?

not too far off for the slope, but the intercept is off, and we are definitely not capturing the quadratic term
